Use `geom_bar` to make a histogram of penguin counts by `year`.

```r
options(repr.plot.width=5, repr.plot.height=2)

p = ggplot(penguins, aes(y=year, fill=species, color=species))
    
p = p+geom_bar(binwidth=5, size=1)+
    labs(x="Number of Penguins", y="Years")+
    scale_color_brewer(palette='Dark2')+
    scale_fill_brewer(palette='Dark2')

p
```

Make a histogram or KDE plot that shows the distribution of body mass in each year. *Use a different theme than the default.*

```r
library(scales)
```

```r
options(repr.plot.width=20, repr.plot.height=15)

p = ggplot(penguins, aes(x=body_mass_g, fill=species, color=species)) + 
    geom_histogram(na.rm = TRUE, binwidth =80, position = 'identity', alpha=0.5)+
    facet_grid(vars(year))+
    labs(y="Number of Penguins", x="Body Mass in Grams", title="Histogram of Penguin Body Mass for Measurements Taken from 2007-2009")+
    scale_color_brewer(palette='Set2')+
    scale_fill_brewer(palette='Set2')+
    scale_x_continuous(labels = comma_format()) +
    theme_minimal() + 
    theme(legend.position = 'top')
p
```

Is there a difference between the body mass of male and female penguins? Use a scatter plot with jitter to compare the masses of male and female penguins. **Facet by `species` to compare within each species-group.**

```r
options(repr.plot.width=7, repr.plot.height=6)

p = ggplot(penguins, aes(x=body_mass_g, y=sex)) + 
    geom_point(na.rm = TRUE, position = position_jitter(height = 0.4, width=0.2), shape=21, alpha=0.5, color='#e66101', fill='#e66101')+
    facet_wrap(vars(species))+
    theme(axis.text.x = element_text(angle = 45, hjust = 1))+
    labs(y="Male or Female Penguins", x="Body Mass in Grams")
p
```
